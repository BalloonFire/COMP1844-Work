```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Create graph object
MyGraph = nx.Graph()

# Create Piccadilly Line Graph:
# Add nodes
MyGraph.add_node('Hyde Park Corner', npos=(0, 2), ccn='#0000FF')
MyGraph.add_node('Green Park', npos=(5, 4), ccn='#0000FF')
MyGraph.add_node('Piccadilly Circus', npos=(9, 4), ccn='#0000FF')
MyGraph.add_node('Leicester Square', npos=(12, 6), ccn='#0000FF')
MyGraph.add_node('Covent Garden', npos=(14, 8), ccn='#0000FF')
MyGraph.add_node('Holborn', npos=(18, 10), ccn='#0000FF')

# Add edges
MyGraph.add_edge('Hyde Park Corner', 'Green Park', cce='#0000FF', distance="0.5")
MyGraph.add_edge('Green Park', 'Piccadilly Circus', cce='#0000FF', distance="0.4")
MyGraph.add_edge('Piccadilly Circus', 'Leicester Square', cce='#0000FF', distance="0.3")
MyGraph.add_edge('Leicester Square', 'Covent Garden', cce='#0000FF', distance="0.2")
MyGraph.add_edge('Covent Garden', 'Holborn', cce='#0000FF', distance="0.4")

# Extract attributes from graph to dictionary
pos = nx.get_node_attributes(MyGraph, 'npos')
nodecolour = nx.get_node_attributes(MyGraph, 'ccn')
edgecolour = nx.get_edge_attributes(MyGraph, 'cce')

# Place dictionary values in array
nodearray = nodecolour.values()
edgearray = edgecolour.values()

plt.figure(figsize=(10, 7))

# Display stations name
plt.text(0, 1.5, s='Hyde Park Corner', rotation=0)
plt.text(5.3, 4.2, s='Green Park', rotation=15)
plt.text(9.5, 3.8, s='Piccadilly Circus', rotation=15)
plt.text(12.5, 5.8, s='Leicester Square', rotation=15)
plt.text(14.5, 7.8, s='Covent Garden', rotation=15)
plt.text(18, 10.3, s='Holborn', rotation=0)

# Draw graph nodes and edges
nx.draw_networkx(MyGraph, pos, node_color=nodearray, with_labels=False)
nx.draw_networkx_edges(MyGraph, pos, edge_color=edgearray)

# Draw edge distances labels 
edge_labels = nx.get_edge_attributes(MyGraph, 'distance')
edge_labels = {k: f"{v} m" for k, v in edge_labels.items()}
nx.draw_networkx_edge_labels(MyGraph, pos, edge_labels=edge_labels)

# Add legend (key)
blue_line = Line2D([0], [1], color='#0000FF', lw=2, label='Piccadilly')
plt.legend(handles=[blue_line], loc='lower right', title="Key", frameon=True)

# Visualize graph
plt.title('London Public Transport Network')
plt.show()
```
